```
#LINEAR REGRESSION
 
from sklearn import linear_model
 
x=df.iloc[:,0:1].values
y=df.iloc[:,6].values
 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
 
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
LinReg.fit(x_train,y_train)
 
plt.scatter(x_train,y_train, color='green')
plt.plot(x_train,LinReg.predict(x_train),color='blue')
plt.title('Linear Regression')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```
#POLYNOMILA REGRESSION
 
from sklearn.preprocessing import PolynomialFeatures
 
poly=PolynomialFeatures(degree=10)
x_poly=poly.fit_transform(x_train)
x_poly
 
polyreg =LinearRegression ()
polyreg.fit(x_poly,y_train)
 
#rsquare
 
y_predict_slr = LinReg.predict(x_test)
 
from sklearn import metrics
r_square =metrics.r2_score(y_test,y_predict_slr)
print(r_square)
 
Y_predict_slr = polyreg.predict(poly.fit_transform(x_test))
R_square =metrics.r2_score(y_test,Y_predict_slr)
print(R_square)
 
a=LinReg.predict([[2000]])
print (a)
b=polyreg.predict(poly.fit_transform([[2000]]))
print (b)
 
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(y_test,y_predict_slr))
print('MSE:',metrics.mean_squared_error(y_test,y_predict_slr))
print('RMSE:',np.sqrt(metrics.mean_absolute_error(y_test,y_predict_slr)))
```
